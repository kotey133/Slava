```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import io
import scipy.stats as stats
```

```python
date_string = '''
names Estimate Std.Error
(Intercept) 3446.830     81.840
totwrk        -0.169      0.018
age            2.714      1.472
south        102.272     41.925
male          87.108     35.173
smsa         -54.188     33.193
yngkid       -13.051     50.459
marr          31.360     42.263
union         11.866     38.186'''
date_string
```

```python
df = pd.read_csv(io.StringIO(date_string), sep='\s+')
df
```

```python
df['t_nabl'] = np.abs(df['Estimate']/df['Std.Error'])
df
```

```python
alpha = 0.05
t_crit = stats.t.ppf(1-alpha/2, 2,706-9)
t_crit
```

```python
df['Znachimosty'] = df.apply(lambda x: 'znachim' if x['t_nabl']>t_crit else 'ne znachim', axis=1)
df
```
